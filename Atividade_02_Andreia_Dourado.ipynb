```python
def impar(list):
  out = []
  for i in list:
    if i%2!=0:
      out.append(i)
  return out
```

##Questão 02

```python
def primos(list):
  out=[]
  for i in list:
    if i<2:
      i = False
    for j in range(2,i):
      if i%j==0:
        i = False
    if i: out.append(i)
  return out
```

##Questão 03

```python
def elementos(list_1, list_2):
  out=[]
  for i in list_1:
    if i in list_2:
      i = False
    if i: out.append(i)
  return out
```
